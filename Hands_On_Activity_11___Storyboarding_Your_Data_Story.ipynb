**3. Discussions:**

Storyboarding helps plan how your data story will unfold just like scenes in a movie.
It helps you decide what charts, annotations, or visuals are needed to communicate your message effectively.

A typical storyboard has:

Frame 1: Title or question (What are we exploring?)

Frame 2: Key data visualization or comparison

Frame 3: Main insight or conclusion

In Python, you can simulate a storyboard using multiple subplots each representing a scene or frame in your story.

**4. Procedures:**

Storyboard Example – “Sales Growth Over Time”

```python
# Import libraries
import matplotlib.pyplot as plt

# Sample data
years = [2020, 2021, 2022, 2023, 2024]
sales = [100, 150, 200, 300, 450]

# Create storyboard using subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Frame 1: Title / Introduction
axes[0].text(0.5, 0.5, "Our Sales Journey\n(2020–2024)", ha='center', va='center', fontsize=14)
axes[0].axis("off")

# Frame 2: Key Data Visualization
axes[1].plot(years, sales, marker='o', color='steelblue')
axes[1].set_title("Sales Over Time")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Sales (in units)")

# Frame 3: Insight / Message
axes[2].text(0.5, 0.5, "Insight:\nSales grew steadily,\nwith a big jump after 2022!",
             ha='center', va='center', fontsize=12, color='darkgreen')
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

**5. Supplementary Activity:**

1. Choose a small dataset (e.g., from Kaggle or manually created).
Then, create a 3-frame storyboard in Python showing:

Frame 1: Title or question

Frame 2: One simple chart (bar or line)

Frame 3: Main message or conclusion

2. Use matplotlib and label each part clearly.

```python
#Code here

# Import libraries
import matplotlib.pyplot as plt

# Sample dataset: Average yearly temperature (°C)
years = [2020, 2021, 2022, 2023, 2024]
avg_temp = [28.1, 28.4, 28.7, 29.3, 29.8]

# Create a storyboard using subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Frame 1: Title / Question
axes[0].text(0.5, 0.5, "How Has Temperature Changed\nFrom 2020 to 2024?",
             ha='center', va='center', fontsize=14, fontweight='bold', color='darkred')
axes[0].axis("off")

# Frame 2: Simple Chart (Line)
axes[1].plot(years, avg_temp, marker='o', color='orange', linewidth=2)
axes[1].set_title("Average Temperature per Year", fontsize=12)
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Temperature (°C)")
axes[1].grid(True, linestyle='--', alpha=0.5)

# Frame 3: Main Message / Conclusion
axes[2].text(0.5, 0.5,
             "Insight:\nA steady rise in average temperature\nsuggests a warming trend.",
             ha='center', va='center', fontsize=12, color='firebrick')
axes[2].axis("off")

# Adjust layout and show
plt.tight_layout()
plt.show()
```

**6. Conclusion/Learnings/Analysis:**

The storyboard does give a clear indication of a gradual warming trend over the years through a steady rise in average temperature from 2020 to 2024. It could mean that the area or the environment under study is getting hotter little by little, perhaps because of climate change or some other environmental factors.
